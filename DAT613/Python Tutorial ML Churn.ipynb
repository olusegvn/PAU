## Feature Extraction

```python
features = data[['gender','Dependents','MonthlyCharges', 'TotalCharges']]
```

```python
# features
```

```python
# Convert categorical data to numerical
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize the LabelEncoder
le = LabelEncoder()
```

```python
features["gender"] = le.fit_transform(features["gender"])
```

```python
features
```

```python
features["Dependents"] = le.fit_transform(features["Dependents"])
```

```python
features
```

```python
target = le.fit_transform(data["Churn"])
```

```python
target
```

## Split your data in training and testing sets

```python
# Importing the function used for splitting
from sklearn.model_selection import train_test_split
```

```python
feature_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3)
```

```python
feature_train
```

```python
target_train
```

```python
print(feature_train.shape, features_test.shape, target_train.shape, target_test.shape)
```

##  Building the model

```python
# Building a classification model using Suport vector machine
# Import the suport the svm library/function

from sklearn.svm import SVC
```

```python
# Initailize the SVC function
classifier = SVC()
```

```python
feature_train.info()
```

```python
# Convert "TotalCharges" to float for training
feature_train["TotalCharges"] = le.fit_transform(feature_train["TotalCharges"])
```

```python
# Convert "TotalCharges" to float for testing
features_test["TotalCharges"] = le.fit_transform(features_test["TotalCharges"])
```

```python
feature_train
```

```python
# Training the model
model = classifier.fit(feature_train, target_train)
```

```python
# Making predictions
prediction = model.predict(features_test)
```

```python
prediction
```

```python
#Compare actual with predictions using a dictionary
diction = {"Actual  value": target_test, "Predictions": prediction}
```

```python
diction
```

```python
# Convert dictionary to a dataframe
pd.DataFrame(diction)
```

## Evaluate your model

```python
# Import the function for  accuracy
from sklearn.metrics import accuracy_score
```

$\texttt{accuracy}(y, \hat{y}) = \frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples}-1} 1(\hat{y}_i = y_i)$

```python
# Check the accuracy of your model
accuracy_score(target_test, prediction)
```

```python
# Using confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Printing the confusion matrix
print(confusion_matrix(target_test, prediction))
```

```python
cm = confusion_matrix(target_test, prediction, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
```

```python
from sklearn.metrics import classification_report
# target_names = ['class 0', 'class 1']
print(classification_report(target_test, prediction))
```
